```python
import os
import shutil
import random

# Define the path to your dataset folder
dataset_path = '/Users/gk/Documents/semester_6/project_sahu_sir/p5/train'

# Define the path to store the train and validation datasets
train_path = '/Users/gk/Documents/semester_6/project_sahu_sir/p5/train_split/train'
valid_path = '/Users/gk/Documents/semester_6/project_sahu_sir/p5/train_split/valid'

# Define the ratio of train-validation split
train_ratio = 0.9
valid_ratio = 0.1

# Iterate through each folder (label) in the dataset
for label_folder in os.listdir(dataset_path):
    label_folder_path = os.path.join(dataset_path, label_folder)
    if os.path.isdir(label_folder_path):
        # List all images in the label folder
        images = [f for f in os.listdir(label_folder_path) if f.endswith('.JPG') or f.endswith('.png')]
        # Shuffle the images randomly
        random.shuffle(images)
        
        # Calculate the split index for train and validation sets
        train_split_index = int(len(images) * train_ratio)
        valid_split_index = int(len(images) * (train_ratio + valid_ratio))
        
        # Split images into train and validation sets
        train_images = images[:train_split_index]
        valid_images = images[train_split_index:valid_split_index]
        
        # Create train and validation folders if they don't exist
        train_label_path = os.path.join(train_path, label_folder)
        valid_label_path = os.path.join(valid_path, label_folder)
        os.makedirs(train_label_path, exist_ok=True)
        os.makedirs(valid_label_path, exist_ok=True)
        
        # Move or copy images to train and validation folders
        for image in train_images:
            src = os.path.join(label_folder_path, image)
            dest = os.path.join(train_label_path, image)
            shutil.copy(src, dest)  # Use shutil.move if you want to move instead of copy
        
        for image in valid_images:
            src = os.path.join(label_folder_path, image)
            dest = os.path.join(valid_label_path, image)
            shutil.copy(src, dest)  # Use shutil.move if you want to move instead of copy
```
